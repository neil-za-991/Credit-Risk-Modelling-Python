```python
# population stability index (PSI)
# new data
# present data (actual) and past data (expected) distributions of features
# 
# PSI = sigma[(%actualj - %expectedj) x ln(%actualj/%expectedj)]
# 0 and 1
# 0 = no difference
# 1 = absolute difference
# < 0.1 = little to no difference
# 0.1 < PSI < 0.25 = little difference (no action taken)
# PSI > 0.25 = big difference (action taken)
```
